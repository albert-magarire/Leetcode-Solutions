```python
from openai import OpenAI
from google.colab import userdata

# Get the API key from Colab's secrets manager
client = OpenAI(api_key=userdata.get("OPENAI_API_KEY"))

response = client.responses.create(
    model="gpt-5",
    input="return the most optimal java solution for this coding problem, with all code commented inline, and handle all/any possible edge cases: Given a rows x cols binary matrix filled with 0's and 1's, find the largest rectangle containing only 1's and return its area"
)

print(response.output_text)
```
